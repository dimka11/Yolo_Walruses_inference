```python
import numpy as np
import pandas as pd
import json
import os
import glob
import cv2
import matplotlib.pyplot as plt
```

```python
test_image_path = './test_images/DJI_0001 (2).JPG'
test_centers_path =  './coords_center/DJI_0001 (2).JPG.csv'
image = cv2.imread(test_image_path)
centers = pd.read_csv(test_centers_path)
centers.head()
```

```python
for i, row in centers.iterrows():
    cv2.circle(img=image, center = (np.int32(row['x']),np.int32(row['y'])), radius =5, color =(255,0,0), thickness=-1)
plt.figure(figsize=(30, 20))
plt.imshow(image)
```
